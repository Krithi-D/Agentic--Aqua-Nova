```python
import pandas as pd

df = pd.read_csv("water_potability.csv")
pd.set_option("display.max_rows", None)
df
```

```python
df.info()
df.describe()
df.head()
```

```python
df = df.dropna()
```

```python
from sklearn.ensemble import IsolationForest

features = ['ph', 'Turbidity', 'Sulfate']
X = df[features]

model = IsolationForest(contamination=0.1, random_state=42)
model.fit(X)


df['anomaly'] = model.predict(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Turbidity'], df['ph'], c=df['anomaly'], cmap='coolwarm')
plt.xlabel("Turbidity")
plt.ylabel("pH")
plt.title("Anomaly Detection")
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np


df = df.reset_index()
X = df[['index']]
y = df['Turbidity']

reg_model = LinearRegression()
reg_model.fit(X, y)


future_steps = np.array([[len(df) + i] for i in range(1, 6)])
predicted_values = reg_model.predict(future_steps)

print(predicted_values)
```

```python
import joblib
joblib.dump(model, "anomaly_model.pkl")
joblib.dump(reg_model, "forecast_model.pkl")
```

```python
import pandas as pd

df = pd.read_csv("temp, oxygen data set.csv")
df = df.dropna()

df = df.reset_index(drop=True)
df["minutes"] = df.index
```

```python
from sklearn.linear_model import LinearRegression

X_temp = df[["minutes"]]
y_temp = df["Temperature (C)"]

temp_model = LinearRegression()
temp_model.fit(X_temp, y_temp)
```

```python
X_oxy = df[["minutes"]]
y_oxy = df["Dissolved Oxygen (mg/L)"]

oxy_model = LinearRegression()
oxy_model.fit(X_oxy, y_oxy)
```

```python
import numpy as np

last_min = df["minutes"].max()
future_times = np.array([[last_min + i] for i in range(1, 6)])

temp_pred = temp_model.predict(future_times)
oxy_pred = oxy_model.predict(future_times)

for i in range(5):
    print(f"In {i+1} min  Temp: {round(temp_pred[i], 2)} C, Oxygen: {round(oxy_pred[i], 2)} mg/L")
```

```python
import matplotlib.pyplot as plt

# For Temperature
plt.figure(figsize=(6,4))
plt.plot(df["minutes"], df["Temperature (C)"], label="Temp")
plt.plot(range(last_min+1, last_min+6), temp_pred, label="", linestyle="--")
plt.xlabel("Minutes")
plt.ylabel("Temperature (C)")
plt.title("Temperature Prediction")
plt.legend()
plt.grid(True)
plt.show()

# For Dissolved Oxygen
plt.figure(figsize=(6,4))
plt.plot(df["minutes"], df["Dissolved Oxygen (mg/L)"], label="DO")
plt.plot(range(last_min+1, last_min+6), oxy_pred, label="", linestyle="--")
plt.xlabel("Minutes")
plt.ylabel("Dissolved Oxygen (mg/L)")
plt.title("Oxygen Prediction")
plt.legend()
plt.grid(True)
plt.show()
```

```python
output_df = pd.DataFrame({
    "Minute": range(last_min+1, last_min+6),
    "Predicted Temp (C)": temp_pred,
    "Predicted Oxygen (mg/L)": oxy_pred
})

output_df.to_csv("forecast_results.csv", index=False)
```
